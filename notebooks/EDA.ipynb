```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df =pd.read_csv("C:\HR_ANALYTICS\data\HR_Employee_Attrition.csv")
print(df.shape)
df.head()
```

```python
df.info()
df.isnull().sum()
df.duplicated().sum()
```

```python
# How many employees left vs stayed?
df["Attrition"].value_counts()
sns.countplot(x='Attrition',data=df)
plt.title('Attrition Distribution')
plt.show()
```

```python
sns.countplot(x='Department',hue='Attrition',data=df)
plt.title('Attrition by Departement')
plt.xticks(rotation=15)
plt.show()
```

```python
sns.countplot(x='OverTime',hue='Attrition',data=df)
plt.title('Overtime vs Attrition')
plt.show()
```

```python
sns.boxplot(x='Attrition',y='MonthlyIncome',data=df)
plt.title('Monthly Income vs Attrition')
plt.show()
```

```python
numeric_df=df.select_dtypes(include='number')
plt.figure(figsize=(14,10))
sns.heatmap(numeric_df.corr(),annot=False,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
## Key Findings:
#1. Employees who work overtime leave at 3x the rate of those who don't
#2. Sales department has the highest attrition rate (21%)
#3. Employees who left earned on average $2,000/month less than those who stayed
```
